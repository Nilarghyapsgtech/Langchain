Recurssive Character Text Splitter

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import PyPDFLoader
from pathlib import Path

pdf_path=Path('./SRCs')

docs=list(pdf_path.glob('*.pdf'))

loader=PyPDFLoader(docs[0])

docs=loader.load()

splitter=RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=15,
    length_function=len,
    separators=[
        "\n\n",
        "\n",
        " ",
        ".",
        ",",
    ]
)

splits=splitter.split_documents(docs)
print(len(splits))
```

For Splitting Code

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter,Language

python_code='''

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Cannot divide by zero"
    return a / b

def modulus(a, b):
    return a // b

    '''

splitter=RecursiveCharacterTextSplitter.from_language(
    language=Language.PYTHON,
    chunk_size=50,
    chunk_overlap=5,

)

splits=splitter.split_text(python_code)
for index,chunk in enumerate(splits):
    print('*'*25)
    print(f"Chunk{index}")
    print('*'*25)
    print('\n')
    print(chunk)
    print('\n')


```

HTML Recurssive Header Text Splitter

```python
from langchain_text_splitters import HTMLHeaderTextSplitter
from langchain_openai import ChatOpenAI
from pathlib import Path

src_path='SRCs'
html_path=list(Path(src_path).glob('*.html'))[0]
html_content=''
if Path(html_path).exists():
    with open(html_path,'r',encoding='utf-8') as f:
        html_content=f.read()
else:
    print("File Doesnot exists")

html_split_headings=[
    ("h1", "Header 1"),
    ("h2", "Header 2"),
    ("h3", "Header 3"),

]

splitter=HTMLHeaderTextSplitter(headers_to_split_on=html_split_headings)
splitted_text=splitter.split_text(html_content)

for index,chunk in enumerate(splitted_text[:3]):
    print('*'*25)
    print(f"Chunk{index}")
    print('*'*25)
    print('\n')
    print(chunk.page_content)
    print('\n')

```
