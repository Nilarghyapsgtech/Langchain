Embedding of Text 

```python
from langchain_openai import OpenAIEmbeddings

embedding_model=OpenAIEmbeddings(model='text-embedding-3-small')

query="What is life ?"

embedding=embedding_model.embed_query(query)

print(embedding[:3])
```

Embedding of Document

```python
from langchain_openai import OpenAIEmbeddings
from pathlib import Path


txt_path=list(Path('SRCs').glob('*.txt'))[0]
txt_content=''
if txt_path.exists():
    with open(txt_path,'r',encoding='utf-8') as file:
        txt_content=file.read()

embedding_model=OpenAIEmbeddings(model='text-embedding-3-small')
embedding=embedding_model.embed_documents(txt_content)

print(embedding[0][:10])
```

Cosine Similarity

```python
import numpy as np
from langchain_openai import OpenAIEmbeddings

embedding=OpenAIEmbeddings(model='text-embedding-3-small')

def cosine_similarity(vec1,vec2):
    num=np.dot(vec1,vec2)
    denm=np.linalg.norm(vec1)*np.linalg.norm(vec2)
    return num/denm


texts=["I love to visit mountains","I went to Himalayas","I live in Kolkata"]
vectors=[]
for text in texts:
    vectors.append(embedding.embed_query(text))

difference_1_2=cosine_similarity(vectors[0],vectors[1])
difference_2_3=cosine_similarity(vectors[1],vectors[2])
difference_3_1=cosine_similarity(vectors[2],vectors[0])

print(f"The Cosine similarity of Vector1 and Vector2 {difference_1_2}")
print(f"The Cosine similarity of Vector2 and Vector3 {difference_2_3}")
print(f"The Cosine similarity of Vector3 and Vector1 {difference_3_1}")


```
